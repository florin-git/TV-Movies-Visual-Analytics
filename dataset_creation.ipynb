```python
import pandas as pd
from pathlib import Path  
```

```python
dataset_path = "./dataset/movie_without_december.tsv"
```

```python
df = pd.read_csv(dataset_path, sep="\t")
```

```python
# Make titles in lower case
df['title'] = df['title'].str.lower()
```

```python
tv_movies = df['title'].unique()
```

```python
tv_movies.shape
```

```python
kaggle_path = "./dataset/Kaggle/ImdbTitleBasics.csv"
```

```python
df_imdb = pd.read_csv(kaggle_path, sep=",")
```

```python
df_imdb
```

```python
df_imdb.rename(columns = {'originalTitle':'title'}, inplace = True)
df_imdb['title'] = df_imdb['title'].str.lower()
df_imdb['primaryTitle'] = df_imdb['primaryTitle'].str.lower()
```

```python
only_movies = df_imdb[df_imdb['titleType'] == "movie"]
```

```python
only_movies_without_duplicates = only_movies.drop_duplicates(subset=['title'], keep=False)
```

```python
only_movies_without_duplicates
```

```python
df_merge = pd.merge(df, only_movies_without_duplicates, how="inner", on="title") 
```

```python
titles_merge = df_merge['title'].unique()
```

```python
titles_merge.shape
```

```python
rating_path = "./dataset/Kaggle/ImdbTitleRatings.csv"
df_rating  = pd.read_csv(rating_path, sep=",")
df_merge = pd.merge(df_merge,df_rating, how="inner", on="tconst")
df_merge
```

```python
#Remove columns not necessary
df_merge.pop('tconst')
df_merge.pop('isAdult')
df_merge.pop('endYear')
df_merge.pop('numVotes')
df_merge.pop('primaryTitle')
df_merge.pop('titleType')

```

```python
df_merge.rename(columns = {'duration':'duration_with_adv', 'startYear':'year', 'runtimeMinutes':'duration', 'averageRating':'rating', }, inplace = True)
data = df_merge
data
```

```python
number_movies = data.groupby(["channel","month"])["channel"].count()
from pathlib import Path  
filepath = Path('./dataset/Kaggle/number_movies.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
number_movies.to_csv(filepath) 
```

```python
mov = pd.read_csv('./dataset/Kaggle/number_movies.csv', sep=",")
final_df = pd.merge(data, mov, on=["month","channel"])
final_df.rename(columns = {'channel.1':'number_movies'}, inplace = True)

filepath = Path('./dataset/Kaggle/final_df.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
final_df.to_csv(filepath) 
```

```python
final_df = pd.read_csv('./dataset/Kaggle/final_df.csv', sep=",")
# data["duration"] = pd.to_numeric(data["duration"], downcast="float")
mean_duration = final_df.groupby(["channel","month"])["duration"].mean()
filepath = Path('./dataset/Kaggle/mean_duration.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
mean_duration.to_csv(filepath)
mean_duration =  pd.read_csv('./dataset/Kaggle/mean_duration.csv', sep=",")
```

```python
mean_duration_with_adv = final_df.groupby(["channel","month"])["duration_with_adv"].mean()
filepath = Path('./dataset/Kaggle/mean_duration.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
mean_duration_with_adv.to_csv(filepath)
mean_duration_with_adv =  pd.read_csv('./dataset/Kaggle/mean_duration.csv', sep=",")
duration = pd.merge(mean_duration, mean_duration_with_adv, on=["month","channel"])
final_df = pd.merge(final_df, duration, on=["month","channel"])
final_df

```
